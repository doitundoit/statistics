# 복습 문제

```python
import pandas as pd
import pingouin as pg
```

## 두 집단의 평균 비교(t-test)

```python
smokers = pd.read_excel('data/birthsmokers.xlsx')
smokers.head()
```

```python
yes = smokers[smokers.Smoke == 'yes'].Wgt
```

```python
no = smokers[smokers.Smoke == 'no'].Wgt
```

```python
pg.ttest(yes, no)
```

## 분산 분석과 사후 검정

```python
d = pd.read_excel('data/depression.xlsx')
d.head()
```

```python
d.TRT.nunique()
```

```python
pg.homoscedasticity?
```

```python
pg.homoscedasticity(d, group='TRT', dv='y')
```

```python
pg.welch_anova(d, between='TRT', dv='y')
```

```python
pg.pairwise_gameshowell(d, between='TRT', dv='y')
```

## 상관 분석

```python
cancer = pd.read_excel('data/skincancer.xlsx')
cancer.head()
```

```python
p = pg.corr(cancer.Mort, cancer.Lat)
p
```

```python
p['p-val'] < 0.05
```

```python
pg.corr(cancer.Mort, cancer.Long)
```

## 회귀 분석

```python
from statsmodels.formula.api import ols
```

```python
speed = pd.read_excel('data/speed.xlsx')
speed.head()
```

```python
s = ols('dist ~ speed', speed).fit()
s.summary()
```

```python
smokers.head()
```

```python
o = ols('Wgt ~ Gest', smokers).fit()
o.summary()
```

# 회귀분석을 통한 예측

```python
car = pd.read_excel('data/car.xlsx')
car.head()
```

```python
o = ols('price ~ mileage', car).fit()
o.summary()
```

```python
# new_car = pd.DataFrame({'mileage': [10000, 20000]})
new_car = pd.read_excel('data/new_car.xlsx')
o.predict(new_car)
```

```python
o = ols('dist ~ speed', speed).fit()
o.summary()
```
