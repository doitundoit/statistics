# 분산 분석

```python
import pandas as pd
import pingouin as pg
```

```python
hr = pd.read_excel('data/hr.xlsx')
hr.head()
```

```python
p = pg.anova(dv='rating', between='job_level', data=hr)['p-unc']
s = 0.05

p < s
```

# 사후 검정

1. 집단 간 분산이 같은지 검증: pg.homoscedasticity()
2. 같으면: Tukey HSD `pg.pairwise_tukey()`
   다르면: Games-Howell `pg.pairwise_gameshowell()`

```python
# job_level -> rating 
# 집단 간 분산이 같은지 검증(등분산성): equal_var = False 분산이 다름
pg.homoscedasticity(dv='rating', group='job_level', data=hr)
```

```python
# 분산이 다른 경우 Games-Howell : pg.pairwise_gameshowell() -> pval, hedges
pg.pairwise_gameshowell(dv='rating', between='job_level', data=hr)
```

```python
# job_level -> overtime
# 집단 간 분산이 같은지 검증(등분산성): equal_var = True 분산이 같음
pg.homoscedasticity(dv='overtime', group='job_level', data=hr)
```

```python
# 분산이 같을 경우 Tukey HSD : pg.pairwise_tukey() -> p-tukey, hedges
pg.pairwise_tukey(dv='overtime', between='job_level', data=hr)
```

```python
census = pd.read_excel('data/census.xlsx')
census.head()
```

```python
# 귀무가설: 모든 인종의 평균 교육 기간이 같다.
result = pg.anova(dv='education_num', between='race', data=census)
result['p-unc'] < 0.05
# p < 0.05 이므로, 모든 인종의 평균 교육 기간이 다를 수 있다.
# 등분산성 검증하고 
```

```python
# 등분산성 검증: 다름 
pg.homoscedasticity(dv='education_num', group='race', data=census)
```

```python
# 등분산성 검증이 다르면 welch_anova로 분산 분석 실시
pg.welch_anova(dv='education_num', between='race', data=census)
```

```python
# p-unc < 0.05 이므로 사후 검정 실시
# 분산이 다르므로 games-howell
result = pg.pairwise_gameshowell(dv='education_num', between='race', data=census)
result
```

```python
# Amer-Indian-Eskimo 집단과 통계적으로 유의한 차이가 있는 집단

cond = (result['A'] =='Amer-Indian-Eskimo') & (result['pval'] < 0.05)
result[cond]
```

# 상관분석

```python
car = pd.read_excel('data/car.xlsx')
car.head()
```

```python
p = pg.corr(car.price, car.mileage)
# p['p-val'] < 0.05
p
```

```python
import matplotlib.pyplot as plt
plt.plot(car.mileage, car.price, 'o', alpha=0.5)
```
